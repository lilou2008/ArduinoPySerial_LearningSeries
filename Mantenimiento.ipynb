```python

import sqlite3
from fpdf import FPDF
from google.colab import output, files  # <--- AGREGA 'files' aquí
from IPython.display import display, HTML
from ipywidgets import Text, Button, HBox, VBox, Layout, Dropdown, Textarea

# === GENERAR PDF ===
def generar_pdf(datos):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size = 15)
    pdf.cell(200, 10, txt = "INFORME DE MANTENIMIENTO", ln = True, align = 'C')
    pdf.ln(10)

    for k, v in datos.items():
        pdf.set_font("Arial", size = 12)
        pdf.cell(30, 10, txt = k, border = 1)
        pdf.cell(150, 10, txt = str(v), border = 1, ln = True)

    pdf.output("informe_mantenimiento.pdf")
    files.download("informe_mantenimiento.pdf")  # Ya está importada
    print("Descarga iniciada...")
```

```python

import ipywidgets as widgets
from IPython.display import display, clear_output, FileLink, HTML
from ipywidgets import Layout, HBox, VBox
from datetime import datetime
import fpdf
from google.colab import files
!pip install fpdf  # Instala la librería para PDF si no la tienes

# Función para generar PDF
def generar_pdf(b):
    # Datos del formulario
    fecha = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    codigo = herramienta_codigo.value
    actividades = actividades_mantenimiento.value
    estado_seleccionado = estado_equipo.value
    estado_otros = estado_otros_text.value if estado_seleccionado == "Otro" else ""

    # Crear PDF
    pdf = fpdf.FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size = 15)

    # Título
    pdf.cell(200, 10, txt = "FORMULARIO DE MANTENIMIENTO", ln = True, align = 'C')
    pdf.ln(10)  # Salto de línea

    # Fuente para secciones
    pdf.set_font("Arial", size = 12)

    # Fecha y Hora
    pdf.cell(0, 10, txt = f"Fecha/Hora: {fecha}", ln = True, align = 'L')

    # Código de Herramienta
    pdf.cell(0, 10, txt = f"Código Herramienta: {codigo}", ln = True, align = 'L')

    # Actividades de Mantenimiento
    pdf.cell(0, 10, txt = "Actividades de mantenimiento:", ln = True, align = 'L')
    pdf.set_font("Arial", size = 10)  # Fuente más pequeña para texto largo
    pdf.multi_cell(0, 6, txt = actividades, align = 'L')  # Imprime texto multilinea
    pdf.set_font("Arial", size = 12)  # Vuelve a la fuente original
    pdf.ln(5)  # Salto adicional para separación

    # Estado del Equipo
    estado_texto = estado_seleccionado
    if estado_seleccionado == "Otro":
        estado_texto += f": {estado_otros}"
    pdf.cell(0, 10, txt = f"Estado del Equipo: {estado_texto}", ln = True, align = 'L')

    # Guarda el PDF
    pdf.output("mantenimiento.pdf")

    # **VERIFICA GENERACIÓN Y DESCARGA**
    import os
    if os.path.isfile("mantenimiento.pdf"):
        display(HTML("<h3 style='color:blue'>Descargando...</h3>"))
        files.download("mantenimiento.pdf")
        clear_output(wait=True)
    else:
        display(HTML("<h3 style='color:red'>Error: No se generó el PDF</h3>"))

    display(widget_form)
    display(FileLink("mantenimiento.pdf", "Descargar PDF (si no se descargó)"))

# Widgets del formulario
fecha_actual = widgets.Label(
    value=datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    layout=Layout(width='100%')
)

herramienta_codigo = widgets.Text(
    value='',
    placeholder='Código de la herramienta',
    layout=Layout(width='100%'),
    description='Código de la herramienta:',
    disabled=False
)

actividades_mantenimiento = widgets.Textarea(
    value='',
    placeholder='Describa las actividades realizadas',
    layout=Layout(width='100%', height='100px'),
    description='Actividades de mantenimiento:'
)

estado_equipo = widgets.Dropdown(
    options=['Operativo', 'Inoperativo', 'En espera de repuesto', 'Otro'],
    value='Operativo',
    layout=Layout(width='100%'),
    description='Estado del equipo:'
)

estado_otros_text = widgets.Text(
    value='',
    placeholder='Describa el estado',
    layout=Layout(width='100%', visibility='hidden'),
    description='Especifique:'
)

# Control de visibilidad para "Otro"
def mostrar_otros(change):
    if change['new'] == "Otro":
        estado_otros_text.layout.visibility = 'visible'
    else:
        estado_otros_text.layout.visibility = 'hidden'

estado_equipo.observe(mostrar_otros, names='value')

# Botones
boton_pdf = widgets.Button(description='Generar y descargar PDF')
boton_pdf.on_click(generar_pdf)

# **BOTÓN ADICIONAL PARA DESCARGA MANUAL**
boton_descargar = widgets.Button(description='Descargar manualmente')
boton_descargar.on_click(lambda b: files.download("mantenimiento.pdf"))

# Estructura del formulario
widget_form = VBox([
    widgets.HBox([widgets.Label("Fecha/Hora Actual:"), fecha_actual]),
    herramienta_codigo,
    actividades_mantenimiento,
    estado_equipo,
    estado_otros_text,
    boton_pdf,
    boton_descargar  # Nuevo botón
], layout=Layout(display='flex', flex_flow='column', align_items='stretch'))

# Mostrar el formulario
display(widget_form)
```
